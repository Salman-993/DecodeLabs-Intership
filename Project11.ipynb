```python
clean_df = df.copy()

change_log = []

def add_log(change_id, description, impact, status="Resolved"):
    change_log.append({
        "Change ID": change_id,
        "Description": description,
        "Impact": impact,
        "Status": status
    })
```

```python
clean_df.columns = clean_df.columns.str.strip()

add_log(
    "CR001",
    "Column names cleaned by removing extra spaces",
    f"{len(clean_df.columns)} columns checked and standardized"
)
```

```python
before_rows = len(clean_df)

clean_df = clean_df.drop_duplicates()

removed_duplicates = before_rows - len(clean_df)

add_log(
    "CR002",
    "Full duplicate rows removed",
    f"{removed_duplicates} duplicate rows removed"
)

print("Removed full duplicate rows:", removed_duplicates)
```

```python
missing_coupon = clean_df['CouponCode'].isnull().sum()

clean_df['CouponCode'] = clean_df['CouponCode'].fillna('NO_COUPON')

add_log(
    "CR004",
    "Missing CouponCode values filled",
    f"{missing_coupon} missing CouponCode values replaced with NO_COUPON"
)

print("Missing CouponCode fixed:", missing_coupon)
```

```python
clean_df['Date'] = pd.to_datetime(clean_df['Date'], errors='coerce')

invalid_dates = clean_df['Date'].isnull().sum()

print("Invalid dates:", invalid_dates)
```

```python
before_rows = len(clean_df)

clean_df = clean_df.dropna(subset=['Date'])

removed_invalid_dates = before_rows - len(clean_df)

clean_df['Date'] = clean_df['Date'].dt.strftime('%Y-%m-%d')

add_log(
    "CR005",
    "Date column converted to YYYY-MM-DD format",
    f"{removed_invalid_dates} invalid date rows removed"
)

clean_df[['OrderID', 'Date']].head()
```

```python
numeric_cols = ['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']

for col in numeric_cols:
    clean_df[col] = pd.to_numeric(clean_df[col], errors='coerce')

numeric_missing = clean_df[numeric_cols].isnull().sum()
numeric_missing
```

```python
for col in numeric_cols:
    missing_count = clean_df[col].isnull().sum()
    
    if missing_count > 0:
        median_value = clean_df[col].median()
        clean_df[col] = clean_df[col].fillna(median_value)
        
        add_log(
            f"CR_NUM_{col}",
            f"Missing values in {col} filled using median",
            f"{missing_count} missing values filled"
        )
```

```python
text_cols = clean_df.select_dtypes(include='object').columns

for col in text_cols:
    clean_df[col] = clean_df[col].astype(str).str.strip()
    clean_df[col] = clean_df[col].str.replace(r'\s+', ' ', regex=True)

add_log(
    "CR006",
    "Text columns cleaned",
    f"{len(text_cols)} text columns cleaned by trimming spaces and removing extra whitespace"
)
```

```python
clean_df['OrderID'] = clean_df['OrderID'].str.upper()
clean_df['CustomerID'] = clean_df['CustomerID'].str.upper()
clean_df['TrackingNumber'] = clean_df['TrackingNumber'].str.upper()
clean_df['CouponCode'] = clean_df['CouponCode'].str.upper()

clean_df['Product'] = clean_df['Product'].str.title()
clean_df['PaymentMethod'] = clean_df['PaymentMethod'].str.title()
clean_df['OrderStatus'] = clean_df['OrderStatus'].str.title()
clean_df['ReferralSource'] = clean_df['ReferralSource'].str.title()
clean_df['ShippingAddress'] = clean_df['ShippingAddress'].str.title()

add_log(
    "CR007",
    "Text formats standardized",
    "IDs and coupon codes converted to uppercase; category columns converted to title case"
)
```

```python
expected_total = (clean_df['Quantity'] * clean_df['UnitPrice']).round(2)

total_mismatch = (clean_df['TotalPrice'].round(2) != expected_total).sum()

print("TotalPrice mismatches:", total_mismatch)
```

```python
final_missing = clean_df.isnull().sum().sum()
final_full_duplicates = clean_df.duplicated().sum()
final_duplicate_order_ids = clean_df['OrderID'].duplicated().sum()
final_duplicate_tracking = clean_df['TrackingNumber'].duplicated().sum()

date_format_errors = ~clean_df['Date'].astype(str).str.match(r'^\d{4}-\d{2}-\d{2}$')
final_date_errors = date_format_errors.sum()

expected_total = (clean_df['Quantity'] * clean_df['UnitPrice']).round(2)
final_total_mismatch = (clean_df['TotalPrice'].round(2) != expected_total).sum()

print("Final Missing Values:", final_missing)
print("Final Full Duplicate Rows:", final_full_duplicates)
print("Final Duplicate OrderID:", final_duplicate_order_ids)
print("Final Duplicate TrackingNumber:", final_duplicate_tracking)
print("Final Date Format Errors:", final_date_errors)
print("Final TotalPrice Mismatches:", final_total_mismatch)
print("Final Dataset Shape:", clean_df.shape)
```
